```python
import smtplib          # 파이썬 기본 내장
from email.mime.text import MIMEText

smtp = smtplib.SMTP('smtp.gmail.com', 587)   # TLS일 경우 포트 587, SSL은 포트 465
smtp.ehlo()
smtp.starttls()    # TLS 사용시 보안 위해 필요
smtp.login('##지메일 계정 이메일 주소 입력##', '##password##')    # 여기서 password는 지메일 비번이 아닌 IMAP 설정

# 본문 메세지
msg = MIMEText('##이메일 본문 메세지##')     # MIMEText: 간단한 본문 메세지, MIMEMultipart: 첨부파일+메세지 등을 포함하는 복합적인 메세지

msg['Subject'] = '##이메일링 테스트##'
msg['To'] = '##받는 이메일##'              # 여러명에게 보내는 경우 따옴표 안에 콤마 처리(리스트화). 예) 'kim@naver.com, lee@naver.com'
receiver = msg['To'].split(",");
smtp.sendmail('##보내는 이메일##', receiver, msg.as_string())
 
smtp.quit()     # 세션 종료
```

### IMAP 설정하는 방법

1. 지메일 계정 로그인
2. 우측 상단 메뉴를 통해 지메일 설정으로 들어감
3. 상단 '전달 및 POP/IMAP' 카테고리를 선택하고 IMAP 액세스에서 IMAP 사용 선택
4. 변경사항 저장 클릭




### 앱 비밀번호 설정

1. 구글 계정 로그인 상태에서 우측 상단 계정 아이콘 클릭
2. Google 계정 관리 클릭

3. 스크롤 다운하여 앱 비밀번호(보안 내부에 메뉴 존재) 클릭
3. 앱 비밀번호 생성 - 앱선택: 메일, 기기 - 필요기기로 선택
