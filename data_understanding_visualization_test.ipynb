```python
# for using pie-chart, import matplotlob

import matplotlib.pyplot as plt

x = [44, 97, 16]
labels = ['X', 'Y', 'Z']

plt.pie(x=x, labels=labels, autopct='%.2f')
```

```python
plt.pie(x=x, labels=labels, autopct='%.1f%%')
```

```python
#explode : array-like, default None

explodes =(0,0,0.1)

plt.pie(x=x, labels=labels, autopct='%.1f%%',explode=explodes)
plt.title("set explode")
plt.show()
```
